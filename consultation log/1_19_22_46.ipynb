### **1. 建立模板思维**
```python
# 二叉树遍历模板（必须背熟）
# 递归模板
def traverse(root):
    if not root:
        return
    # 前序
    traverse(root.left)
    # 中序
    traverse(root.right)
    # 后序

# 迭代模板（前序）
def preorder_iterative(root):
    if not root:
        return []
    stack, result = [root], []
    while stack:
        node = stack.pop()
        result.append(node.val)
        if node.right: stack.append(node.right)  # 右先入栈
        if node.left: stack.append(node.left)    # 左后入栈
    return result
```
